```python
import PyPDF2
import os
from tkinter.filedialog import askdirectory

# Solicita ao usuário que selecione um diretório
caminho = askdirectory(title="Selecione uma pasta")

# Dicionário para organizar arquivos em subpastas com extensões correspondentes
locais = {
    "PDF": [".pdf"],
    "LAYOUT": [".layout"],
    "GIOTTO": [".gio", ".bak"],
    "IMAGENS": [".jpg", ".png", ".jpeg"],
    "ARQUIVOS DE CORTE": [".csv", ".xls", ".pgmx", ".xcs"],
    "AUTOCAD": [".dwg", ".dxf"],
    "SKETCHUP": [".skp", ".skb"],
}

# Obtém a lista de arquivos no diretório e ordena
lista_arquivos = sorted(os.listdir(caminho))

# Organiza os arquivos em subpastas com base em suas extensões
for arquivo in lista_arquivos:
    # Obtém nome e extensão do arquivo
    name, extensao = os.path.splitext(arquivo)
    extensao = extensao.lower()  # Normaliza para evitar problemas com maiúsculas/minúsculas
    
    # Percorre as categorias do dicionário para classificar os arquivos
    for pasta, extensoes in locais.items():
        if extensao in extensoes:
            destino = os.path.join(caminho, pasta)

            # Cria a pasta se não existir
            if not os.path.exists(destino):
                os.mkdir(destino)

            # Move o arquivo para a subpasta correspondente
            os.rename(os.path.join(caminho, arquivo), os.path.join(destino, arquivo))
            break  # Encerra o loop após encontrar uma correspondência

# Lista novamente os arquivos, agora dentro da pasta de PDFs
caminho_pdf = os.path.join(caminho, "PDF")
pdfs = sorted(f for f in os.listdir(caminho_pdf) if f.endswith(".pdf"))

# Inicializa o objeto merger para combinar arquivos PDF
merger = PyPDF2.PdfMerger()

# Adiciona os PDFs ao merger
for pdf in pdfs:
    merger.append(os.path.join(caminho_pdf, pdf))

# Salva o PDF combinado
saida_pdf = os.path.join(caminho, "PDF", "#pdf_final.pdf")
merger.write(saida_pdf)
merger.close()

print(f"PDF combinado salvo em: {saida_pdf}")
```
