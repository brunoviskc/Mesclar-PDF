```python
# Importa a biblioteca PyPDF2 para manipulação de arquivos PDF
import PyPDF2
# Importa o módulo os para interagir com o sistema operacional (listar arquivos, etc.)
import os
# Importa a função askdirectory do tkinter para abrir uma janela de seleção de pasta
from tkinter.filedialog import askdirectory
# Abre uma janela de diálogo para o usuário selecionar uma pasta
# Retorna o caminho da pasta selecionada como string
caminho = askdirectory(title="Seleceione uma pasta")
# Cria um objeto PdfMerger que será usado para combinar múltiplos PDFs em um só
merger = PyPDF2.PdfMerger()
# Lista todos os arquivos e pastas dentro do diretório selecionado
# Retorna uma lista com os nomes dos arquivos
lista_arquivos = os.listdir(caminho)
# Ordena a lista de arquivos em ordem alfabética/numérica
# Isso garante que os PDFs sejam mesclados na ordem correta
lista_arquivos.sort()
# Percorre cada arquivo na lista ordenada
for arquivo in lista_arquivos:
    # Verifica se o arquivo é um PDF (contém ".pdf" no nome)
    if ".pdf" in arquivo:
        # Adiciona o arquivo PDF ao merger usando o caminho completo
        # append() adiciona todas as páginas do PDF ao documento final
        merger.append(f"{caminho}/{arquivo}")

# Escreve/salva o PDF mesclado no mesmo diretório
# O arquivo final será nomeado "#PDF_FINAL.pdf"
# O "#" no início garante que fique no topo da lista de arquivos
merger.write(f"{caminho}/#PDF_FINAL.pdf")
```
